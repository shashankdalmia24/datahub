```python
hubs = listdir("../data/grafana")
if ".DS_Store" in hubs:
    hubs.remove(".DS_Store")
hubs  
```

```python
memory = []
for hub in hubs:
    l = listdir("../data/grafana/" + hub)
    name = [i for i in l if 'memory' in i][0]
    temp = pd.read_csv("../data/grafana/" + hub + "/" + name)
    
    assert temp.columns[0] == 'Time'
    assert temp.columns[1] == 'Value'
    
    temp = temp.rename(columns = {"Value": name[:-4]})
    temp["Time"] = pd.to_datetime(temp["Time"])
    memory.append(temp)
```

```python
for i in range(1, len(memory)):
    try:
        assert memory[0]["Time"].where(memory[0]["Time"].values == memory[i]["Time"].values).notna().sum() == 733
    except:
        print(hubs[i])
```

```python
memory[3]
```

offset Data102 by an hour

```python
memory[3]["Time"] = memory[3]["Time"].apply(lambda x: x - timedelta(hours=1))
```

```python
memory = [i.loc[:731] for i in memory]
[i.shape for i in memory]
```

```python
memory = [i.loc[:731] for i in memory]
for i in range(1, len(memory)):
    try:
        assert memory[0]["Time"].where(memory[0]["Time"].values == memory[i]["Time"].values).notna().sum() == 732
    except:
        print(hubs[i])
```

We can now merge the data on time axis. Note, for a given date data is recorded on multiple timestamps. So we only have value on that timestamp as a flow variable.

```python
memory_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Time'],
                                            how='outer'), memory)
memory_merged["date"] = memory_merged.Time.dt.date
memory_merged["weekday"] = memory_merged.Time.dt.weekday
memory_merged
```

Since we care about the proportion of costs attributed to hub, if we take the max usage from timestamps of a hub for a day, we will still get an unbiased estimate for proportion.

```python
memory_daily_raw = memory_merged.groupby("date").max().drop(columns = ["Time", "weekday"])
memory_daily_raw["daily_total"] = memory_daily_raw.sum(axis = 1)
memory_daily_raw = memory_daily_raw.reset_index()
memory_daily_raw["date"] = pd.to_datetime(memory_daily_raw["date"]).dt.date
memory_daily_raw
```

The hub values are the max recorded usage on that day used as a proxy for the actual usage of the day. the Daily total is the sum of all hubs for the day.

In the following table, i divide a hubs value for a given day to calculate the proportion for the total usage for the day.

```python
memory_daily = memory_merged.groupby("date").max().drop(columns = ["Time", "weekday"])
daily_total = memory_daily.sum(axis = 1)
for col in memory_daily.columns:
    memory_daily[col] = memory_daily[col]/daily_total
memory_daily["daily_total"] = daily_total.values
memory_daily = memory_daily.reset_index()
memory_daily["date"] = pd.to_datetime(memory_daily["date"])
memory_daily
```

```python
cols = ['prob140_memory', 'datahub_memory', 'data100_memory',
       'data102_memory', 'rhub_memory', 'eecs_memory', 'biology_memory']
plt.figure(figsize = (16,9))
for col in cols:
    plt.plot(memory_daily["date"].values, memory_daily[col].values, label = col)
plt.legend()
plt.xlabel("Date")
plt.ylabel("Share of daily total")
plt.title("Share of daily total memory usage");
```

```python
memory_daily["month"] = memory_daily.date.dt.month
memory_daily["weekday"] = memory_daily.date.dt.weekday
memory_daily
```

Over here, i calcuate `monthly_sum`. This is the sum of the `daily_total` usage for the month.

```python
memory_daily = memory_daily.join(memory_daily.groupby('month')['daily_total'].sum(), on='month', rsuffix='_r')
memory_daily = memory_daily.rename(columns = {"daily_total_r": "monthly_sum"})
memory_daily
```

Here I take the share a day of the usage for the month.

```python
memory_daily["share_of_month"] = memory_daily.daily_total / memory_daily.monthly_sum
memory_daily
```

Usage data for the months of October and April are for half of the month each. So, I am remvoing those from analysis. (We also dont have costs for April)

```python
temp = memory_daily[memory_daily.month.isin([11,2,3])]
plt.figure(figsize = (16,9))
plt.plot(temp["date"].values, temp["share_of_month"].values)
plt.xlabel("Date")
plt.ylabel("Share of month")
plt.title("Days's share of Months total for Nov, Feb and March");
```

```python
feat = ["month", "weekday" , "monthly_sum" , "share_of_month"]
for f in feat:
    memory_daily_raw[f] = memory_daily[f]
memory_daily_raw
```

```python
mapping = pd.DataFrame(zip([10,11,12,1,2,3], RAM_costs[9:]))
mapping.columns = ["month", "monthy_RAM_cost"]
mapping
```

```python
usage_costs = memory_daily.merge(mapping)
usage_costs = usage_costs[usage_costs.month != 10]
usage_costs
```

In the above table, I have `monthy_RAM_cost` for a month. Multiply it by `share_of_month` to get the costs for a day. Multiply it by `prob140_memory` to get the share of monthly costs attributed to prob140 hub for that day. Similarly for all hubs, as calcuated below,

```python
cols = ['prob140_memory', 'datahub_memory', 'data100_memory',
       'data102_memory', 'rhub_memory', 'eecs_memory', 'biology_memory']
for col in cols:
    usage_costs[col] = usage_costs[col] * usage_costs.share_of_month * usage_costs.monthy_RAM_cost
usage_costs
```

```python
cols = ['prob140_memory', 'datahub_memory', 'data100_memory',
       'data102_memory', 'rhub_memory', 'eecs_memory', 'biology_memory']
plt.figure(figsize = (16,9))
for col in cols:
    plt.plot(usage_costs["date"].values, usage_costs[col].values, label = col)
plt.legend()
plt.xlabel("Date")
plt.ylabel("Estimated RAM costs for day");
```

Note: Data 100 costs surge in march twice. 

Datahub costs are high and periodic.

Use the following function to plot costs for any hub for any months.

```python
def hub_moth_RAM_cost(hubs, month):
    temp = usage_costs[usage_costs.month.isin(month)]
    plt.figure(figsize = (16,9))
    for col in hubs:
        plt.plot(temp["date"].values, temp[col].values, label = col)
    plt.legend()
    plt.xlabel("Date")
    plt.ylabel("Estimated RAM costs for day");
```

```python
hub_moth_RAM_cost(["data100_memory"], [3])
```
